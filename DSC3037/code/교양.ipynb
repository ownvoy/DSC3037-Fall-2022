## 교양 코드

### 2019년 1학기, 2학기

```python
import json, pandas as pd
with open(r'C:\Users\77sjy\Desktop\sqlProject\data\\01교양\\2019학년도 1학기.json',encoding='UTF8') as f:
    js = json.loads(f.read()) ## json 라이브러리 이용
df2 = pd.DataFrame(js)
df2 = pd.read_json(r'C:\Users\77sjy\Desktop\sqlProject\data\\01교양\\2019학년도 1학기.json',encoding='UTF8') ## pd.read_json 이용
```

```python
df2
```

```python
# '담기' 제거
df2_list = df2['course'].to_list()

for each in df2_list:
    for i in range(len(each)):
        if each[i]== '담기':
            idx = i
            each.remove(each[idx])
            break

# df2_ list를 dataframe으로
df2['course'] = df2_list
```

```python
df2
```

```python
df2['학수번호']=df2['course'].str.get(0)
df2['이수구분']=df2['course'].str.get(1)
df2['과목이름']=df2['course'].str.get(2)
df2['과목영어이름']=df2['course'].str.get(3)
df2['교수명']=df2['course'].str.get(4)
df2['캠퍼스']=df2['course'].str.get(5)
df2['학위과정']=df2['course'].str.get(6)
df2['시간']=df2['course'].str.get(7)
df2['수업형태']=df2['course'].str.get(8)
df2['학점(시수)']=df2['course'].str.get(9)
df2=df2.drop(['course'], axis='columns')
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
df2
```
